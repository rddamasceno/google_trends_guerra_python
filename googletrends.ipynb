```python
import pandas as pd
import matplotlib.pyplot as plt
from pytrends.request import TrendReq

# Criar objeto TrendReq. Português e ajuste da time zone
pytrends = TrendReq(hl='pt-BR', tz=360)

# Configurar os termos de pesquisa e período de tempo
termos = [ 'Guerra', 'Hamas', 'Israel']

# Construir o payload de solicitação
pytrends.build_payload(termos, timeframe = '2023-01-01 2023-12-31', geo = 'BR')

# Obter os dados dos interesses por região
dados = pytrends.interest_over_time()


# Plota o gráfico de linha historica
dados.plot(figsize=(20, 6), marker='o')
plt.xlabel('Data')
plt.ylabel('Interesse em 2023')
plt.title('Buscas no Google Trends por Guerra, Hamas e Israel no Brasil')
plt.show() 
```

```python
# Define o termo de pesquisa
termo_pesquisa = 'Guerra'

# Configura os parâmetros da pesquisa
pytrends.build_payload([termo_pesquisa], timeframe='today 1-m', geo='BR')

# Obtém os dados do Google Trends para o Brasil
dados_brasil = pytrends.interest_over_time()

# Calcula a variação percentual das pesquisas
var_percentual = (dados_brasil[termo_pesquisa][-1] - dados_brasil[termo_pesquisa][0]) / dados_brasil[termo_pesquisa][0] * 100

print(f"Variação percentual das pesquisas por {termo_pesquisa} no Brasil no último mês: {var_percentual:.2f}%")
```
